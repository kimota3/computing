### .plot_surface 로 그리기

```python
# 2변수 함수 그래프

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d


ax3 = plt.axes(projection='3d') #3차원 축


x = np.linspace(-3, 3, 1000) # 정의역 정의(x)
y = np.linspace(-3, 3, 1000) # 정의역 정의(y)


X, Y = np.meshgrid(x, y) # 정의역의 meshgrid

fXY=(2*X**2+Y**2)*np.e**(-X**2-Y**2) #함수(치역)

ax3.plot_surface(X, Y, fXY) #그래프 만들기


plt.show()
```

## 그래프 꾸미기

- 그래프 크기
- 보이는 각도 조절  
- 축 이름 표시
- plot_surface 옵션(xx,yy,temp,rstride = 1,cstride = 1,alpha = 0.3,color = 'blue',edgecolor = 'black')


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

plt.figure(figsize=(7,7)) # 크기

ax3 = plt.axes(projection='3d') 


x = np.linspace(-3, 3, 101) 
y = np.linspace(-3, 3, 101) 


X, Y = np.meshgrid(x, y) 

fXY=(2*X**2+Y**2)*np.e**(-X**2-Y**2) 

ax3.plot_surface(X,Y,fXY,rstride = 2,cstride = 20,alpha = 0.4,color = 'gray',edgecolor = 'blue')
 #(X,Y,fXY,열선,행선(1:전부표시, 2:한칸씩 띄워서표시), 투명도, 색깔, 색깔)
    
ax3.xaxis.set_tick_params(labelsize=5) #x축 눈금
ax3.yaxis.set_tick_params(labelsize=10)
ax3.zaxis.set_tick_params(labelsize=15)
ax3.set_xlabel(r'$x$', fontsize =10) #x축
ax3.set_ylabel(r'$y$', fontsize =10)
ax3.set_zlabel(r'$z$', fontsize =20)
#ax3.set_xlabel('x', fontsize =20) # x축
#ax3.set_ylabel('y', fontsize =20)  # y축

ax3.view_init(15,60) # 상하회전각도 ( 0 - 옆에서 본 각도, 90 - 위에서 본 각도)
                     # 좌우회전각도 ( 양수 - 시계방향 회전, 음수 - 시계반대방향 회전)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

plt.figure(figsize=(7,7))

ax = plt.axes(projection='3d')
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)
ax.set_xlabel(r'$x$', size =12)
ax.set_ylabel(r'$y$', size =12)
ax.set_zlabel(r'$z$', size =12)

def f(x,y):
    z= -(x+1)*(x)*(x-1)*(x-2)-(y+1)*(y)*(y-1)*(y-2)
    #z= -(x+1)*(x)*(x-1)*(x-2)-(y+1)*(y)
    return z

## 정의역정의(x1, x2)
x= np.linspace(-1.2, 2) 
y= np.linspace(-1, 2)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1,X2)
#Z = 50 * (X2 - X1 **2) **2 + (2 - X1)**2

ax.scatter3D(X1, X2, Z, marker='*', color='blue')

plt.show()
```

```python
import sympy as sp
import numpy as np 
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
#f = x**2 + y**2 +x*y
k =10*x**2 - 10*y**2 
#j = x**2 - y**2

#def k(x,y):
 #   z = x**2 + y**2 + x*y - sp.sin(x)*5
  #  return z

k_by_x = sp.diff(k, x)
k_by_y = sp.diff(k, y)

# make meshgrid with xs, ys 
n = 21
x_3 = np.linspace(-0.5, 0.5,n)
y_3 = np.linspace(-0.5, 0.5, n)
X_3, Y_3= np.meshgrid(x_3, y_3)

# make zs 
#zs = [float(k.subs(x, xv).subs(y, yv)) for xv, yv in zip(X_3.ravel(), Y_3.ravel())]
#zs = np.array(zs).reshape(sample_size, sample_size)
Z_3 = [float(k.subs(x,xv).subs(y,yv)) for xv, yv in zip(X_3.ravel(), Y_3.ravel())]
Z_3=np.array(Z_3).reshape(n,n)
#print(Z_3)

fig=plt.figure(figsize=(16, 8))

#fig=plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X_3, Y_3, Z_3, cmap='viridis')
ax.view_init(10,70) 
plt.show()

#ax=fig.add_subplot(121)
plt.contour(X_3, Y_3, Z_3, 50, levels = np.arange(-1, 1, 0.2), cmap=plt.cm.rainbow)

# gradient

X_4, Y_4 = np.meshgrid( np.linspace(-0.5,0.5, 11), np.linspace(-0.5, 0.5, 11) )

XG = [float(k_by_x.subs(x, xx).subs(y, yy)) for xx, yy in zip(X_4.ravel(), Y_4.ravel())]

YG = [float(k_by_y.subs(x, xx).subs(y, yy)) for xx, yy in zip(X_4.ravel(), Y_4.ravel())]

plt.quiver(X_4, Y_4, XG, YG, width=0.005, scale=100, color='gray')
plt.show()
```

```python
import sympy as sp
import numpy as np 
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
k = x**2 + y**2 + x*y - sp.cos(x)*5

k_by_x = sp.diff(k, x)
k_by_y = sp.diff(k, y)

# make meshgrid with xs, ys 
n=50
x_3 = np.linspace(-10, 10, n)
y_3 = np.linspace(-10, 10, n)
X_3, Y_3= np.meshgrid(x_3, y_3)

# make zs 
#zs = [float(k.subs(x, xv).subs(y, yv)) for xv, yv in zip(X_3.ravel(), Y_3.ravel())]
#zs = np.array(zs).reshape(sample_size, sample_size)
Z_3 = [float(k.subs(x,xv).subs(y,yv)) for xv, yv in zip(X_3.ravel(), Y_3.ravel())]
Z_3=np.array(Z_3).reshape(n,n)
#print(Z_3)

fig=plt.figure(figsize=(8, 6))

#fig=plt.figure()
#ax = fig.add_subplot(121, projection='3d')
#ax.plot_surface(X_3, Y_3, Z_3, cmap='viridis')
#ax.view_init(10,70) 
#plt.show()

#ax=fig.add_subplot(121)
plt.contour(X_3, Y_3, Z_3, 50, levels = np.logspace(-1.0, 2.0, 20), cmap=plt.cm.rainbow)

# gradient

X_4, Y_4 = np.meshgrid( np.linspace(-10, 10, 10), np.linspace(-10, 10, 10) )

XG = [float(k_by_x.subs(x, xx).subs(y, yy)) for xx, yy in zip(X_4.ravel(), Y_4.ravel())]

YG = [float(k_by_y.subs(x, xx).subs(y, yy)) for xx, yy in zip(X_4.ravel(), Y_4.ravel())]

plt.quiver(X_4, Y_4, XG, YG, width=0.005, scale=500, color='red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 목적함수 f(x) = x1^2 + x2^2
def f1(x1, x2):
    return x1 ** 2 + x2
x1 = np.linspace(-3, 3, 100)
x2 = np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x1, x2)
Y = f1(X1, X2)

# 등식 제한조건 방정식   x1**2 + x2**2 = 1
x2_g = np.sqrt(+1 - x1**2)
x3_g = -np.sqrt(+1 - x1**2)



fig=plt.figure(figsize=(6, 6))
CC=plt.contour(X1, X2, Y, colors="gray", levels=[-1,0.5, 5/4,2, 8])
plt.clabel(CC,inline=2,fontsize = 9)

plt.colorbar
plt.plot(x1, x2_g, 'g-')
plt.plot(x1, x3_g, 'g-')

#plt.plot([], [], 'rP')
#plt.plot([], [], 'ro', ms=10)

#plt.xlim(-3, 1)
#plt.ylim(-1, 1)
#plt.xticks(np.linspace(-3, 3, 9))
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
#plt.title(" ")
plt.show()
```
